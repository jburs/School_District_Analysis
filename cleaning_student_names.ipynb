```python
#import dependencies
import pandas as pd
```

```python
student_data_path = "Resources/students_complete.csv"

student_data_df = pd.read_csv(student_data_path)
student_data_df.head()

#to do: remove all profesoinal prefixes and suffixes (ex. Dr.)
```

```python
#put student names into a list
student_names = student_data_df["student_name"].tolist()
#student_names
```

```python
#new list for names to be fixed
students_to_fix =[]

for name in student_names:
    if len(name.split()) >= 3:
        students_to_fix.append(name)

len(students_to_fix)
```

```python
#add the prefixes <= 4 to a new list (miss is longest prefix)
prefixes =[]

for name in students_to_fix:
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])

#convert to set to get unique items
print(set(prefixes), len(set(prefixes)))
        
```

```python
#add the suffix <= 3 to a new list
suffixes =[]

for name in students_to_fix:
    if len(name.split()[-1]) <= 3:
        suffixes.append(name.split()[-1])
        
#convert to set to get unique items
print(set(suffixes), len(set(suffixes)))
```

```python
#to be removed:
#prefix: Dr., Mr., Mrs., Ms., Miss
#suffix: MD, DDS, DVM, phD, Ph.D
```

```python
prefixes_suffixes = ["Dr. ", "Mr. ", "Mrs.", "Ms. ", "Miss ", 
                     " MD", " DDS", " DVM", " PhD", " Ph.D"]

for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word, "")
    
        
```

```python
#confirm replacements and cleaned names

student_names = student_data_df["student_name"]

students_fixed = []

for name in student_names:
    if len(name.split()) >= 3:
        students_fixed.append(name)
        
len(students_fixed)
```
